```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
columnNames = ["sepalHeight","sepalLenght","petalHeight","petalLenght","classification"] # Creating table columns' names
dataFrame = pd.read_csv("/content/iris.data",names=columnNames) # Reading .csv file
dataFrame["sepalArea"] = dataFrame["sepalHeight"] * dataFrame["sepalLenght"]
dataFrame["petalArea"] = dataFrame["petalHeight"] * dataFrame["petalLenght"]  # Adding two more columns
independentVariables = dataFrame.columns.difference(["classification"],sort=False)
#independentVariables
X = dataFrame[independentVariables].values # Independent variable upper case X
y = dataFrame["classification"] # Dependent variable lower case y
objClassify = DecisionTreeClassifier(criterion="gini",max_depth=2)
objClassify.fit(X,y) # "Feeding" the AI
```

After reading and storing the .csv file, we will display it using IPython.display library.

```python
from IPython.display import Image  
from sklearn import tree
import pydotplus

# Create DOT data
dot_data = tree.export_graphviz(objClassify, out_file=None, 
                                feature_names=independentVariables,  
                                class_names=dataFrame.classification.unique())

# Draw graph
graph = pydotplus.graph_from_dot_data(dot_data)  

# Show graph
Image(graph.create_png())
```
